```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# ==============================
# DATA
# ==============================
np.random.seed(42)
X = np.random.randint(0,100,(10000,2))
y = X[:,0] + X[:,1]
print(X[:5],y[:5])

X = X/100.0
y = y/200.0

# ==============================
# MODEL (16,8,1)
# ==============================
model = keras.Sequential([
    layers.Dense(16,activation='relu',input_shape=(2,)),
    layers.Dense(8,activation='relu'),
    layers.Dense(1)
])

model.compile(
    optimizer='adam',
    loss='mse',
    metrics=['mae']
)

model.summary()

# ==============================
# TRAIN
# ==============================
history = model.fit(
    X,y,epochs=20,batch_size=32,validation_split=0.2
)

# ==============================
# TEST (your existing test)
# ==============================
x_test = X[0:1]
y_true = y[0:1]

y_pred = model.predict(x_test)

print("Pred (scaled):", y_pred[0][0])
print("True (scaled):", y_true[0])
print("Pred (real):", y_pred[0][0] * 200)
print("True (real):", y_true[0] * 200)

```
